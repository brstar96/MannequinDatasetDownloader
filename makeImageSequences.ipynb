## Make Image Sequences using Meta Files
`MannequinChallenge` dataset contains lots of `.txt` files including youtube video url and frame information calculated in microsecond. ([Check this official Dataset site](https://google.github.io/mannequinchallenge/www/download.html))

In this notebook, we will calculate frame number and get single image using frame number. 

### Test on single `.txt` file
With below functions, you can get various informations of video file from scratch. 

```python
import cv2

_capture = cv2.VideoCapture('D:/MannequinChallenge/MannequinChallenge_Videos_Refined/videos/aaxX6usSqBA.mp4')
_meta_txt = 'D:\MannequinChallenge\meta_files\train/'

length = int(_capture.get(cv2.CAP_PROP_FRAME_COUNT))
width = int(_capture.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(_capture.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = _capture.get(cv2.CAP_PROP_FPS)

print('length : ', length)
print('width : ', width)
print('height : ', height)
print('fps : ', fps)
```

```python
import math 
import matplotlib.pyplot as plt

video = cv2.VideoCapture('D:/MannequinChallenge/MannequinChallenge_Videos_Original/train/__D0l4m8_w8.mp4')
f = open("D:/MannequinChallenge/meta_files/train/00abf2d3644c4b04.txt", 'r')
lines = f.readlines()
for line in lines:
    row_str = line.replace('\n', '') 
    if not 'https' in line:
        microsecond_info = int(row_str.split(' ')[0])
#         print('microsecond: {}, corresponding frame_number: {}'.format(int(microsecond_info), math.floor(round(int(microsecond_info)/1000000, 3)*29.97)))
        
        frame_number = math.floor(round(int(microsecond_info)/1000000, 3)*29.97)
        print(frame_number)
        video.set(cv2.CAP_PROP_POS_FRAMES, frame_number-1)
        ret, frame = video.read() # Read the frame
        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        plt.imshow(frame)
        plt.show()
        video
        
f.close()
```
